```python
def create_x_and_y(image_size, target, grid_size, num_classes):
    
    x = tf.zeros((1, image_size, image_size, 3))
    y = tf.zeros((grid_size, grid_size, 3, 5 + num_classes))

    
    boxes = target['boxes']
    labels = target['labels']

    for box, label in zip(boxes, labels):
        
        x_center = ((box[0] + box[2]) / 2) / image_size
        y_center = ((box[1] + box[3]) / 2) / image_size
        box_width = (box[2] - box[0]) / image_size
        box_height = (box[3] - box[1]) / image_size

        
        grid_x = int(x_center * grid_size)
        grid_y = int(y_center * grid_size)

        
        x_center_cell = (x_center * grid_size) - grid_x
        y_center_cell = (y_center * grid_size) - grid_y

        
        y[grid_y, grid_x, 0] = x_center_cell
        y[grid_y, grid_x, 1] = y_center_cell
        y[grid_y, grid_x, 2] = box_width
        y[grid_y, grid_x, 3] = box_height
        y[grid_y, grid_x, 4] = 1  
        y[grid_y, grid_x, 5 + label] = 1  

    return x, y
```

```python
from PIL import Image
import os
import json


# images_directory = "your_images_directory_path_here"

labels_directory = "hold"


# images = []


annotations = []


max_json_files = 50
json_files_processed = 0
import numpy as np
grid_size = 8



for folder in os.listdir(labels_directory):
    folder_path = os.path.join(labels_directory, folder)
    
    
    if os.path.isdir(folder_path):
        
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            
            
            if file.lower().endswith('.json') and 'pre_disaster' not in file:
                if json_files_processed >= max_json_files:
                    break  
                
                annotation_now = []
                y = np.zeros((grid_size, grid_size, 9))
                with open(file_path, 'r') as f:
                    json_data = json.load(f)
                    
                    
                    for item in json_data['features']['xy']:
                        properties = item['properties']
                        subtype = properties['subtype']
                        wkt = item['wkt']
                        if(subtype == 'un-classified'):
                            continue
                        
                        coordinates = wkt[wkt.find("((") + 2:wkt.find("))")]
                        coordinates = coordinates.split(", ")
                        coordinates = [tuple(map(float, coord.split(" "))) for coord in coordinates]
                        
                        
                        x_values = [coord[0] for coord in coordinates]
                        y_values = [coord[1] for coord in coordinates]
                    
                        
                        
                        x_center = ((x_values[0] + x_values[2]) / 2) / image_sizes[0][0]
                        y_center = ((y_values[1] + y_values[3]) / 2) / image_sizes[0][1]
                        box_width = (x_values[2] - x_values[0]) / image_sizes[0][0]
                        box_height = (y_values[3] - y_values[1]) / image_sizes[0][1]

                        grid_size = 8

                        # print(y_center)
                        grid_x = int(x_center * grid_size)
                        grid_y = int(y_center * grid_size)

                        
                        x_center_cell = (x_center * grid_size) - grid_x
                        y_center_cell = (y_center * grid_size) - grid_y

                        class_mapping = {'no-damage': 0, 'major-damage': 1, 'minor-damage': 2, 'destroyed': 3}

                        y[grid_y, grid_x, 0] = x_center_cell
                        y[grid_y, grid_x, 1] = y_center_cell
                        y[grid_y, grid_x, 2] = box_width
                        y[grid_y, grid_x, 3] = box_height
                        y[grid_y, grid_x, 4] = 1  
                        y[grid_y, grid_x, 5 + class_mapping[subtype]] = 1 
                        

                        
                    annotations.append(y)
                    json_files_processed += 1  





# print(images)
# print(annotations)
```
